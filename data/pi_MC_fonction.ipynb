# Détermination du nombre $\pi$ par Monte-Carlo

D'abord on importe les modules

```python
import random
import math

import matplotlib.pyplot as plt
```

On définit une fonction qui calcul $\pi$ par Monte-Carlo en prenant en argument le nombre d'itérations.

```python
def calc_pi_MC(nb_iterations):
    # On initialise notre compteur et on fixe la taille du graphe.
    counts = 0
    plt.figure(figsize=(5, 5))
    # On boucle nb_iterations.
    for i in range(0, nb_iterations):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        r = math.sqrt(x*x + y*y)
        if r <= 1:
            # Dans le cercle.
            counts += 1
            plt.scatter(x, y, color="red")
        else:
            # Hors du cercle.
            plt.scatter(x, y, color="blue")

    print("Nombre de points dans le carré : {}".format(nb_iterations))
    print("Nombre de points dans le cercle : {}".format(counts))
    pi_MC = counts/nb_iterations * 4
    print("Évaluation de pi : {:.6f}".format(pi_MC))
    print("Valeur de pi : {:.6f}".format(math.pi))
    print("Différence : {:.6f}".format(pi_MC - math.pi))
```

Maintenant que la fonction est définie, il ne reste plus qu'à l'appeler avec le nombre d'itérations souhaitées :-) !

```python
calc_pi_MC(100)
```

On recommence avec 1000 itérations, sans avoir à ré-écrire tout le code du Monte-Carlo :-) !

```python
calc_pi_MC(1000)
```

```python
calc_pi_MC(10000)
```
